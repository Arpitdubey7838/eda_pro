```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load iris dataset
df = sns.load_dataset("iris")

# Plot 1-D scatter plot
plt.scatter(df[df.species=="setosa"]["petal_length"], [0]*50, label="setosa")
plt.scatter(df[df.species=="versicolor"]["petal_length"], [0]*50, label="versicolor")
plt.scatter(df[df.species=="virginica"]["petal_length"], [0]*50, label="virginica")

plt.xlabel("Petal length")
plt.yticks([])   # hide y-axis (since it's always 0)
plt.legend()
plt.title("1-D Scatter Plot of Petal Length")
plt.show()

```

```python
# 1-D scatter plot of petal-length
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load iris dataset
iris = sns.load_dataset("iris")

# Split dataset by species
iris_setosa = iris.loc[iris["species"] == "setosa"]
iris_virginica = iris.loc[iris["species"] == "virginica"]
iris_versicolor = iris.loc[iris["species"] == "versicolor"]

# 1-D scatter plot
plt.plot(iris_setosa["petal_length"], np.zeros_like(iris_setosa["petal_length"]), 'o', label="setosa")
plt.plot(iris_versicolor["petal_length"], np.zeros_like(iris_versicolor["petal_length"]), 'o', label="versicolor")
plt.plot(iris_virginica["petal_length"], np.zeros_like(iris_virginica["petal_length"]), 'o', label="virginica")

plt.xlabel("Petal length")
plt.legend()
plt.title("1-D Scatter Plot of Petal Length")
plt.show()
```

# pdf and cdf

```python
#pdf-probability density function (area of that point )
#cdf -commulative density function (sum of previous data area)
counts, bin_edges = np.histogram(df['petal_length'], bins=10, density=True)

pdf = counts / (sum(counts))
print(pdf)
print(bin_edges)

# compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)


```

# 9.9 -> median , percentile , quantile , IQR , MAD

```python
# median(x) = mean (of x50 +x51) = 50th percentile
print(np.median(df[df['species']=="setosa"]["petal_length"]))
print(np.median(df[df['species']=="versicolor"]["petal_length"]))
print(np.median(df[df['species']=="virginica"]["petal_length"]))
print(np.median(np.append(df[df['species']=="virginica"]["petal_length"],50)))

```

```python
# 25th , 50th , 75th , 100th = Quantiles 
# 25th - first quantile , 50th 2nd quantile , 75th 3rd quantile, 100th 4th quantile
print(np.percentile(df[df['species']=="setosa"]["petal_length"] ,np.arange(0,100,25)))  # 0th , 25th , 50th ,75th
print(np.percentile(df[df['species']=="versicolor"]["petal_length"] ,np.arange(0,100,25)))
print(np.percentile(df[df['species']=="virginica"]["petal_length"] ,np.arange(0,100,25)))
```

```python
#90th percentile 
print(np.percentile(df[df['species']=="setosa"]["petal_length"],90))  # percentile mean how many you lead out of 100 
print(np.percentile(df[df['species']=="versicolor"]["petal_length"],90))
print(np.percentile(df[df['species']=="virginica"]["petal_length"],90))
```

```python
# MAD - median absolute deviation 
from statsmodels import robust 
print("*/ median absolute deviation*/")
print(robust.mad(df[df['species']=="setosa"]["petal_length"]))
print(robust.mad(df[df['species']=="versicolor"]["petal_length"]))
print(robust.mad(df[df['species']=="virginica"]["petal_length"]))
```

### median --> mean 
### MAD --> std deviation
### IQR - inter quartile range == 75th percentile - 25th perccentile

# 9.12 - BOX PLOT AND WHISKERS

```python
# Box plot with whiskers : another method of visualization the 1D scatter plot 
# note -: in the plot below  a technique call " inter quartile range " is used in whiskers in the plot below do not correspond to the min and max value.
sns.boxplot(x='species' , y='petal_length',data = df)
plt.show()

# whiskers = 1.5 * IQR
```

# 9.13 violin plots 

```python
# a violin plot combine the benefits of the previous two plots and simplifies them
# denser region of the data are fatter, and sparser ones thinner in a violin plot

sns.violinplot(x="species" , y="petal_length" , data=df )  #we use the plt.figure(figsize=(width,height)) instead of size , beacuse in violinplot size show erroe
plt.figure(figsize=(0,8))
plt.show()
# size is used in sns.stripplot or sns.swarmplot (to control marker size), not in violinplot.
# For controlling the plot size, always use plt.figure(figsize=(...)).
```
